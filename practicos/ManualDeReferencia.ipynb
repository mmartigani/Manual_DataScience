```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.aemet.es/es/eltiempo/observacion/ultimosdatos?k=and&w=0&datos=det&x=h24&f=temperatura"
```

```python
html_content = requests.get(url).text
html_content
```

```python
soup = BeautifulSoup(html_content)
soup
```

```python
print(soup.prettify())
```

```python
soup.title.text
```

```python
weather_table = soup.find("table", attrs={"class": "width100 tabla_dinamica tabla_responsive"})
weather_table
```

```python
table_header = weather_table.thead
header_row = table_header.tr

col_names = list()

for th in header_row.find_all("th"):
    col_names.append(th.text.replace("\n", "").strip("\xa0"))
    
col_names
```

```python
table_body = weather_table.tbody
table_body
```

```python
len(table_body.find_all("tr"))
```

```python
for trow in table_body.find_all("tr"):
    for td in trow.find_all("td"):
        print(td)
        print("\n")
```

```python
table_rows = []

for trow in table_body.find_all("tr"):
    row = []
    for td in trow.find_all("td"):
        row.append(td.text)
        
    table_rows.append(row)
    
table_rows
```

```python
pd.DataFrame(table_rows, columns = col_names)
```
